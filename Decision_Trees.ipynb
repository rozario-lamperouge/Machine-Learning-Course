```
import pandas as pd

# Convert X to a pandas DataFrame
df = pd.DataFrame(X, columns=['height', 'shoe_size'])
df['gender'] = Y

df.head()
```

```
import seaborn as sns
import matplotlib.pyplot as plt

# Scatterplot
sns.scatterplot(x='height', y='shoe_size', hue='gender',data=df)
plt.title("Height vs Shoe Size")
plt.xlabel("Height")
plt.ylabel("Shoe Size")
plt.show()
```
