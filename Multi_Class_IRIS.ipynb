```
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
```

## **Plotting**

```
df = pd.read_csv("Machine-Learning-Course/iris.csv")
```

```
df.head()
```

```
ax = sns.scatterplot(x="petal.length", y="petal.width",hue="Variety",size = "sepal.width",data=df)
plt.show()
```

## **Data Preparation**

```
df = np.array(df)

X = df[:,0:4]
Y = df[:,4]
```

```
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=7)
```

## **Decision Trees**

```
#splitter = best, random , max_features = int, auto, log, none
model = DecisionTreeClassifier(random_state=84)

print("[INFO] training model...")
model.fit(X_train, Y_train)

print("[INFO] evaluating...")
predictions = model.predict(X_test)
print(classification_report(Y_test, predictions))
```

## **Random Forest**

```
# Training
clf = RandomForestClassifier(n_estimators=10, random_state=42)

print('[INFO] training model...')
clf.fit(X_train,Y_train)

print('[INFO] evaluating...')
predict = clf.predict(X_test)
print(classification_report(Y_test,predict))
```

## **K Nearest Neighbors**

```
model = KNeighborsClassifier(n_neighbors=9)
print("[INFO] training model...")
model.fit(X_train, Y_train)


print("[INFO] evaluating...")
predictions = model.predict(X_test)
print(classification_report(Y_test, predictions))
```

## **SVM**

```
model = SVC(kernel="rbf",C=100)
print("[INFO] training model...")
model.fit(X_train, Y_train)


print("[INFO] evaluating...")
predictions = model.predict(X_test)
print(classification_report(Y_test, predictions))
```

## **Logistic Regression**

```
model = LogisticRegression(max_iter=15)#,solver = 'sag')

print("[INFO] training model...")
model.fit(X_train, Y_train)

print("[INFO] evaluating...")
predictions = model.predict(X_test)
print(classification_report(Y_test, predictions))
```
