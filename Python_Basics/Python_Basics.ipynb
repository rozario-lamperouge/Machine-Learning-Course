## **Variables**

A variable is a name that stores a value in memory.
It acts like a label for data you want to use or modify later.

```python
Pi = 3.14
print(Pi)
```

**Variable Naming Rules**

‚úÖ Allowed:

- Can start with a letter or underscore (_)

- Can contain letters, digits, and underscores

- Case-sensitive (age, Age, and AGE are different)

‚ùå Not allowed:

- No spaces ‚Üí user name ‚ùå

- No special characters ‚Üí user-name ‚ùå

- Cannot start with a digit ‚Üí 1st_name ‚ùå

- Cannot use reserved keywords ‚Üí for, while, class, etc.

```python
user_age = 25     # ‚úÖ valid
UserAge = 25      # ‚úÖ valid (different variable)
user-age = 25     # ‚ùå invalid (hyphen not allowed)
```

```python
# Remove values from the list
l.pop()
print(l)
```

## **Dictionary**

```python
D = {1:"richard", 2:"john", 3:"peter"}
print(type(D))
```

```python
# Dictionary - Key:Value Pairs

d = {"one":"Earth",
     "two":"Mars",
     "three":"Venus"}

print(type(d))
```

```python
# Dictionary[Key] => Value
print(d["two"])
```

```python
# Adding new items

d["two"] = "mars"
d["four"] = "jupyter"
print(d)
```

```python
# Deleting items

d.pop('four')
print(d)
```

```python
for k in d:
  print(k, d[k])
```

```python
marks = {"Alice": 85, "Bob": 78, "Charlie": 92}

for name, score in marks.items():
    print(name, "scored", score)
```
