## **1) NumPy: Numerical Python**

```python
a = [1,2,3,4]
type(a)
```

```python
# Efficient arrays for math & scientific computing.
import numpy as np

# Create arrays
a = np.array([1, 2, 3, 4, 5])
print("a:", a)
print(type(a))
```

```python
# 0 to 10 step 2
b = np.arange(0, 10, 2)
print("b:", b)
```

```python
# 5 numbers between 0 and 1
c = np.linspace(0, 1, 10)
print("c:", c)
```

```python
l1 = [1,2,3]
l2 = [4,5,6]

l1 * 2
```

```python
# Vectorized operations
print("a * 2:", a * 2)
print("a + 2:", a + 2)
print("a + b:", a + b)
print("a x b:", a * b)
```

```python
print(a, a.shape)
```

```python
[1,2,3]
[4,5,6] -> 2x3

transpose-
[1,4]
[2,5]
[3,6] -> 3x2

[7,8,9]
[10,11,12] -> 2x3
```

```python
matrix = np.array([[1,2,3],[4,5,6]])
transpose = matrix.T

matrix2 = np.array([[7,8,9],[10,11,12]])

print("matrix:\n", matrix, matrix.shape)
print("matrix2:\n", matrix2, matrix2.shape)
print("transpose:\n", transpose, transpose.shape)

print(matrix * transpose)
```

```python
# 2D arrays
matrix = np.array([[1,2,3],[4,5,6]])

# print shape
print("Matrix shape:", matrix.shape)

# print matrix
print("matrix:\n", matrix)

# transpose of matrix
print("Transpose:\n", matrix.T)
print("Transpose shape:", matrix.T.shape)
```

## **2) Pandas: Data Analysis**

Tabular data with rows & columns (like Excel, SQL).

```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
```

**Here are some of the important fields (columns) in the dataset:**

- PassengerId: unique ID for each passenger

- Survived: whether the passenger survived (0 = No, 1 = Yes)

- Pclass: passenger class (1 = first, 2 = second, 3 = third)

- Name: the name of the passenger

- Sex: gender of the passenger (male/female)

- Age: age of the passenger in years

- SibSp: number of siblings/spouses aboard

- Parch: number of parents/children aboard

- Ticket: ticket number

- Fare: the fare the passenger paid

- Cabin: cabin number (many missing)

- Embarked: port of embarkation (C = Cherbourg, Q = Queenstown, S = Southampton)

```python
# read the csv as a pandas dataframe
df = pd.read_csv(url)
df.head()
```

**1) Inspect the Data**

```python
df.shape
```

```python
df.info()
```

```python
df.head()        # first 5 rows
df.tail()        # last 5 rows
df.shape         # (rows, columns)
df.info()        # structure + dtypes
df.describe()    # stats for numerical columns
```

**2. Select Columns**

```python
subset =  df[["Name", "Age", "Sex"]]
subset.head()
```

```python
df["Name"]        # single column
df[["Name", "Age", "Sex"]]   # multiple columns
```

**3. Filtering Rows**

```python
df[df["Age"] > 50]          # passengers older than 50
```

```python
df[df["Sex"] == "female"]   # all female passengers
```

```python
df[(df["Age"] > 30) & (df["Sex"] == "male")]
```

**4. Handling Missing Values**

```python
df.isnull().sum() # returns list of null values in each column
```

```python
Agemedian = df["Age"].median()
print(Agemedian)
```

```python
df["Age"].fillna(Agemedian, inplace=True)
```

```python
# drop rows missing 'Embarked'
df.dropna(subset=["Embarked"], inplace=True)
```

```python
df.shape
```

**5. Sorting**

```python
df.sort_values("Age", ascending =  False).head()
```

```python
df.head()
```

**6. Adding New Columns**

```python
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1

df.head()
```

**7. Export Cleaned Data**

```python
df.to_csv("cleaned_titanic.csv", index=False)
```

## **3) Matplotlib: Data Visualization**

Matplotlib is Python‚Äôs most common plotting library.
It lets you create graphs, charts, and even display images.

We use it to:

- Understand data patterns

- Visualize model performance

- Inspect images (Computer Vision)

- Plot statistics, distributions, trends

```python
import matplotlib.pyplot as plt
import numpy as np
```

**Basic Line Plot**



```python
x = np.linspace(0, 10, 100)   # 100 values from 0 to 10
y = np.sin(x)

print(x)
print(y)
```

```python
plt.plot(x, y)
plt.show()
```

**Adding Titles, Label, Grid, Legend**

```python
plt.plot(x, y, label="sin(x)", color="blue")

plt.title("Sine Wave")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.legend()
plt.show()
```

**Scatter Plot**

```python
plt.scatter(x,y)
```

```python
x = np.random.rand(50)
y = 2*x + 0.1*np.random.randn(50)

print(x)
print(y)
```

```python
x = np.random.rand(50)
y = 2*x + 0.1*np.random.randn(50)

plt.scatter(x, y, color="purple")
plt.title("Scatter Plot Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

**Bar Chart**

```python
names = ["Alice", "Bob", "Charlie"]
scores = [85, 90, 78]

plt.bar(names, scores, color="orange")
plt.title("Student Scores")
plt.xlabel("Names")
plt.ylabel("Score")
plt.show()
```

**Histogram**

```python
data = np.random.randn(1000)
print(data)
```

```python
data = np.random.randn(1000)  # 1000 normally distributed values

plt.hist(data, bins=100, color="gray", edgecolor="black")
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

## **4) Seaborn: Statistical Visualization**

High-level, prettier plots built on Matplotlib.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load sample dataset
tips = sns.load_dataset("tips")
print(tips.head())
```

```python
# Histogram
# KDE = smooth curve showing distribution shape.

sns.histplot(tips["total_bill"], kde=True)
plt.show()
```

```python
# Scatterplot
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Bill vs Tip")
plt.show()
```

```python
# adding categories

sns.scatterplot(
    x="total_bill",
    y="tip",
    hue="sex",
    style="time",
    data=tips
)
plt.title("Bill vs Tip (Grouped)")
plt.show()
```

```python
# count plot

sns.countplot(x="day", data=tips)
plt.title("Count of Visits by Day")
plt.show()
```

```python
# Boxplot
sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
```
